"""
arange(1, 11)  生成1~11包前不包后的数字，reshape为2行5列的数组
"""

```python
import torch
src = torch.arange(1, 11).reshape((2, 5))   # shape(2, 5)
src
```

```python
index = torch.tensor([[1, 1, 2, 0, 0]])  # 2维张量 shape(1, 5)
index
```

```python
torch.zeros(3, 5, dtype=src.dtype).scatter_(0, index, src)
# 将src中的所有值按照index中的索引值，写入到self中的指定的index位置
```

```python
torch.full((2, 4), 2.).scatter_(1, torch.tensor([[2], [3]]), 1.23, reduce='multiply')
```

```python
torch.full((2, 4), 2.).scatter_(1, torch.tensor([[2],[3]]), 1.23, reduce='add')
```
